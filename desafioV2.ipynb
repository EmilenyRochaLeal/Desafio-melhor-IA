```python
import aisuite as ai
from typing import List
from typing import Dict

client = ai.Client()

def aks(message: str, model: str, system_message: str = "Voc√™ √© um assistente inteligente que responde em portugu√™s.") -> str:
    """
    Envia uma mensagem para um modelo de IA e retorna a resposta.
    """
    client = ai.Client()
    messages = [
        {"role": "system", "content": system_message},
        {"role": "user", "content": message},
    ]

    response = client.chat.completions.create(model=model, messages=messages)
    return response.choices[0].message.content


def answer(questions: List[str], models: List[str]) -> List[List[str]]:
    """
    Obt√©m respostas de diferentes modelos de IA para cada pergunta e retorna uma lista de listas com as respostas.
    """
    all_responses = []

    for q_index, question in enumerate(questions):
        print(f"\n##### Pergunta {q_index + 1}: {question} #####\n")
        responses = []
        for m_index, model in enumerate(models):
            print(f'::::: Modelo {m_index + 1}: {model} :::::')
            response = aks(question, model)
            print(response)
            responses.append(response)
            print('--------------\n')
        all_responses.append(responses)

    return all_responses

def auto_comparison(question: str, responses: List[str], model: str) -> str:
    """
    Envia as respostas obtidas para o pr√≥prio modelo de IA avali√°-las e ranque√°-las com base em crit√©rios definidos.
    """
    # Criando o prompt de avalia√ß√£o
    respostas_formatadas = "\n".join([f"{i + 1}) {resposta}" for i, resposta in enumerate(responses)])

    prompt_avaliacao = f"""
    Considere a seguinte pergunta: "{question}"

    Aqui est√£o as respostas fornecidas por diferentes modelos:
    {respostas_formatadas}

    Por favor, ranqueie essas respostas com base nos seguintes crit√©rios:
    - Clareza e Coer√™ncia
    - Precis√£o da Informa√ß√£o
    - Criatividade ou Profundidade
    - Consist√™ncia Gramatical

    D√™ qual a melhor resposta.
    """

    # Enviando o prompt para o modelo
    avaliacao = aks(prompt_avaliacao, model)
    return avaliacao
```

```python
def main():
    models = [
        'groq:deepseek-r1-distill-llama-70b',
        'anthropic:claude-3-5-sonnet-20241022',
        'groq:llama-3.3-70b-versatile',
    ]

    questions = [
        'O que √© Engenharia de software aplicada?',
        'O que √© uma fun√ß√£o em programa√ß√£o?',
        'Quais os tipos de intelig√™ncia artificial?',
        'Como o c√©rebro realmente aprende?',
        'Porque Oeiras deixou de ser a capital do Piau√≠?',
    ]

    # Obter respostas
    all_responses = answer(questions, models)

    # üìù Autoavalia√ß√£o assistida por IA
    for q_index, question in enumerate(questions):
        print(f"\nüìù Autoavalia√ß√£o para a pergunta: {question}")
        for model in models:
            print(f"\nü§ñ Avalia√ß√£o do modelo {model}:")
            resultado_avaliacao = auto_comparison(question, all_responses[q_index], model)
            print(f"üìä Resultado da avalia√ß√£o:\n{resultado_avaliacao}")
            print("====================================================")


if __name__ == "__main__":
    main()
```
